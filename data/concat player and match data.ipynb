```python
import pandas as pd
import numpy as np
```

```python
df_player = pd.read_csv("./match_player_data.csv")
df_match = pd.read_csv("./match_data.csv")
```

```python
df_player.head()
```

```python
df_match.head()
```

```python
# drop overlapping columns from player df
to_drop = []
for col_name in df_player.columns:
    if col_name in df_match.columns:
        to_drop.append(col_name)

df_player.drop(columns=to_drop, inplace=True)
```

```python
merged_raw = df_match.join(df_player)
```

```python
merged_raw.head()
```

```python
merged_raw.columns
```

```python
# drop unnecessary columns
merged_raw.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
merged_raw.head()
```

```python
# split into tm and opp team
df_tm = merged_raw.iloc[::2]
df_opp = merged_raw.iloc[1::2]
```

```python
df_tm.reset_index(inplace=True, drop=True)
df_opp.reset_index(inplace=True, drop=True)
```

```python
# drop team name columns for the opp df to prevent duplication
df_opp.drop(columns=['Tm', 'Opp'], inplace=True)
```

```python
for col_name in df_opp.columns:
    df_opp.rename(columns={col_name: col_name + "_opp"}, inplace=True)
```

```python
df_opp.columns
```

```python
# drop unnecessary columns from df_opp
df_opp.drop(columns=['Date_opp', 'Result_opp'], inplace=True)
```

```python
merged_final = df_tm.join(df_opp)
```

```python
merged_final.columns
```

```python
# save cleaned data to csv
merged_final.to_csv("./combined_match_data.csv")
```
